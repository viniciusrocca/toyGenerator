```python
#Caso a distribuicao escolhida for a da massa invariante
import time
t0 = time.time()
#Lendo o arquivo
p_Lab = np.genfromtxt('Quadrimomentos.csv', delimiter=",", skip_header =1,usecols =(0, 1, 2, 3, 4, 5, 6, 7))

massa_invariante = []

#Calculo da massa invariante
for i in p_Lab:
    pA = np.array(i[:4])
    pB = np.array(i[4:])
    pTotal = pA + pB
    x = pTotal[:1]**2 - np.dot(pTotal[1:],pTotal[1:])
    #x = np.dot(pTotal[1:],pTotal[1:])
    massa_invariante.append(np.sqrt(x))

massa_invariante = np.array(massa_invariante)

#Criando os histogramas
plt.hist(massa_invariante, bins = 25,
         range = (massa_invariante.min()*0.8,massa_invariante.max()*1.2))

plt.xlabel('M$\gamma$$\gamma$ (GeV)')
plt.ylabel('Eventos')
plt.show()
```

```python
#Caso a distribuicao escolhida for a da massa invariante
import time
t0 = time.time()
#Lendo o arquivo
p_Lab = np.genfromtxt('Quadrimomentos.csv', delimiter=",", skip_header =1,usecols =(0, 1, 2, 3, 4, 5, 6, 7, 12, 13, 14, 15))

massa_invariante = []

#Calculo da massa invariante
for i in p_Lab:
    pA = np.array(i[:4])
    pB = np.array(i[4:8])
    pY = np.array(i[8:])
    pTotal = pA + pB + pY
    x = pTotal[:1]**2 - np.dot(pTotal[1:],pTotal[1:])
    #x = np.dot(pTotal[1:],pTotal[1:])
    massa_invariante.append(np.sqrt(x))

massa_invariante = np.array(massa_invariante)

#Criando os histogramas
plt.hist(massa_invariante, bins = 25,
         range = (massa_invariante.min()*0.8,massa_invariante.max()*1.2), color = 'green')

plt.xlabel('Mq$\gamma$$\gamma$ (GeV)')
plt.ylabel('Eventos')
plt.show()
```

```python
#Caso distribuicao da velocidade de X
t0 = time.time()
#Lendo o arquivo
p_X = np.genfromtxt('Quadrimomentos.csv', delimiter=",", skip_header =1, usecols =(8,9,10,11))

velocidade = []

#Calculo da velocidade
for i in p_X:
    x = np.dot(i[1:],i[1:])
    velocidade.append(np.sqrt(x)/i[0])

velocidade = np.array(velocidade)

#Criando os histogramas
plt.hist(velocidade, bins = 25, histtype='step',
         range = (velocidade.min()*0.8,velocidade.max()*1.2), color = 'green')
plt.title('Mdulo da velocidade de X')
plt.xlabel('Velocidade (c)')
plt.ylabel('Frequencia')
plt.show()
```

```python
#Caso distribuicao da rapidity de X

#Lendo o arquivo
pX = np.genfromtxt('Quadrimomentos.csv', delimiter=",", skip_header =1, usecols =(8, 9, 10, 11))

rap_x = []

#Calculo da rapidity X
for i in pX:
    p_X = np.array(i[:4])
    theta = np.arccos(p_X[3]/(np.sqrt(p_X[1]**2 + p_X[2]**2 + p_X[3]**2)))
    rapidity = np.log(np.tan(theta/2))
    rap_x.append(rapidity)
                
rap_x = np.array(rap_x)

#Criando o histograma
plt.hist(rap_x, bins = 25, histtype='step',
         range = (rap_x.min()*0.8,rap_x.max()*1.2), color = 'red')
plt.title('Pseudorapidity de X')
plt.xlabel('Pseudorapidity')
plt.ylabel('Frequencia')
plt.show()                     

```

```python
#Lendo o arquivo
p_Lab = np.genfromtxt('Quadrimomentos.csv', delimiter=",", skip_header =1,usecols =(0, 4))

e_X = []

#Calculo da massa invariante
for i in p_Lab:
    e_A = np.array(i[0])
    e_B = np.array(i[1])
    e_X.append(e_A + e_B)

e_X = np.array(e_X)

#Criando os histogramas
plt.hist(e_X, bins = 25, histtype='step',
         range = (e_X.min()*0.8,e_X.max()*1.2))
plt.title('eX')
plt.xlabel('Massa (GeV)')
plt.ylabel('Frequencia')
plt.show()
```
